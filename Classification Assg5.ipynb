```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn import tree
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import math

df=pd.read_csv('dermatology_csv.csv')
#x=df.iloc[:,0:34]
#y=df.iloc[:,34]
df.head()

```

```python
#logistic regression

#handling missing values
array=df.values
X=array[:,0:34]
y=array[:,34]
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
imp = IterativeImputer(max_iter=10, random_state=0)
imp.fit(X)
X=np.round(imp.transform(X))

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)

#normalize the feature
scaler = StandardScaler()
X_train=scaler.fit_transform(X_train)

X_test=scaler.fit_transform(X_test)

accuracy=[]
names=[]


logistic_regression=LogisticRegression(solver='lbfgs',multi_class='multinomial',max_iter=1000)

logistic_regression.fit(X_train,y_train)
logistic_regression_predict=logistic_regression.predict(X_test)
# logistic_acc=logistic_regression.score(X,y)
# print('accuracy of logistic regression : {:.3f}%\n'.format(logistic_acc*100))

print('classification Report :\n',classification_report(y_test,logistic_regression_predict))

confusion=confusion_matrix(y_test,logistic_regression_predict)
print(confusion)
```

```python
#Decision tree

#handling missing values
array=df.values
X=array[:,0:34]
y=array[:,34]
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
imp = IterativeImputer(max_iter=10, random_state=0)
imp.fit(X)
X=np.round(imp.transform(X))

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)

decision_tree=DecisionTreeClassifier(criterion='gini')

decision_tree.fit(X_train,y_train)
decision_tree_predict=decision_tree.predict(X_test)
#decision_acc=decision_tree.score(x,y)


print('classification Report :\n',classification_report(y_test,decision_tree_predict))

confusion=confusion_matrix(y_test,decision_tree_predict)
print(confusion)


names.append('Decision tree')
#accuracy.append(decision_acc)
```

```python
#Support Vector Machine
#handling missing values
array=df.values
X=array[:,0:34]
y=array[:,34]
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
imp = IterativeImputer(max_iter=10, random_state=0)
imp.fit(X)
X=np.round(imp.transform(X))

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1)

#normalize the feature
scaler = StandardScaler()
X_train=scaler.fit_transform(X_train)

X_test=scaler.fit_transform(X_test)

accuracy=[]
names=[]


svm=SVC( kernel='linear',max_iter=-1)

svm.fit(X_train,y_train)
svm_predict=svm.predict(X_test)

#svm_accuracy=svm.score(x,y)



print('classification Report :\n',classification_report(y_test,svm_predict))

confusion=confusion_matrix(y_test,svm_predict)
print(confusion)


#accuracy.append(decision_acc)
```
